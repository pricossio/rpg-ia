# Entrenamiento de la IA: RPG Tactics
Este notebook demuestra el entrenamiento de los modelos de Machine Learning (Árboles de Decisión y K-Means) basados en la telemetría exportada de la partida.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
from sklearn.tree import DecisionTreeClassifier, plot_tree
from sklearn.decomposition import PCA

# INSTRUCCIÓN: Sube tu archivo 'dataset_jugadas.csv' en el panel de la izquierda (archivos) de Colab antes de ejecutar esta celda.
df = pd.read_csv('dataset_jugadas.csv')
df.head()
```

```python
X = df[['HP_H1', 'HP_H2', 'HP_H3', 'HP_IA1', 'HP_IA2', 'HP_IA3', 'Clase_Atacante']]
y = df['ID_Objetivo']

# Entrenar Árbol de Decisión
arbol = DecisionTreeClassifier(max_depth=3, random_state=42)
arbol.fit(X, y)

# Entrenar K-Means
kmeans = KMeans(n_clusters=2, random_state=42, n_init='auto')
labels = kmeans.fit_predict(X)

print("Modelos de Machine Learning entrenados con éxito en la nube de Google Colab.")
```

```python
# GRAFICAR K-MEANS
pca = PCA(n_components=2)
X_pca = pca.fit_transform(X)

plt.figure(figsize=(10, 6))
scatter = sns.scatterplot(x=X_pca[:, 0], y=X_pca[:, 1], hue=labels, palette=['#e74c3c', '#3498db'], s=150, edgecolor='black')
plt.title("Clustering K-Means: Agresivo vs Conservador", fontsize=14, fontweight='bold')
plt.show()
```

```python
# GRAFICAR ÁRBOL DE DECISIÓN
plt.figure(figsize=(20, 10))
plot_tree(arbol, 
          feature_names=['Vida H1', 'Vida H2', 'Vida H3', 'Vida IA1', 'Vida IA2', 'Vida IA3', 'Clase'], 
          class_names=[f'Obj {c}' for c in arbol.classes_],
          filled=True, 
          rounded=True, 
          fontsize=10)
plt.title("Árbol de Decisión de la Inteligencia Artificial", fontsize=18, fontweight='bold')
plt.show()
```
